```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 'Mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 그대로 사용
target = df1['판정']

# 테스트 크기를 0.2에서 0.5까지 반복하면서 모델 학습 및 평가
test_sizes = [0.2, 0.3, 0.4, 0.5]

for test_size in test_sizes:
    # 트레인 테스트 스플릿 (이상치 제거를 포함)
    X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=test_size, random_state=42)

    # IQR을 사용하여 이상치 제거
    Q1 = X_train.quantile(0.25)
    Q3 = X_train.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 이상치 제거
    X_train_no_outliers = X_train[(X_train >= lower_bound) & (X_train <= upper_bound)].dropna()
    y_train_no_outliers = y_train.loc[X_train_no_outliers.index]

    # 결정 트리 모델 생성 및 학습
    clf = DecisionTreeClassifier(random_state=42)
    clf.fit(X_train_no_outliers, y_train_no_outliers)

    # 테스트 데이터에 대한 예측
    y_pred = clf.predict(X_test)

    # 정확도 출력
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nTest Size: {test_size}")
    print("Accuracy:", accuracy)

    # 분류 보고서 출력
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # 혼동 행렬 출력
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(conf_matrix)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, accuracy_score, recall_score, precision_score, confusion_matrix, roc_auc_score
from IPython.display import display

# 'mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 0과 1로 변환
df1['판정'] = df1['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

# 타겟 변수 설정
target = df1['판정']

# 트레인 테스트 스플릿
X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=0.4, random_state=42)

# 로지스틱 회귀 모델 생성 및 학습
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# ROC 곡선 생성
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
roc_df = pd.DataFrame({'FPR': fpr, 'TPR': tpr, 'Threshold': thresholds})
roc_df['Threshold'] = roc_df['Threshold'].round(2)

# AUC 계산
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

# 선택된 임계값 범위 설정 (인덱스 2~8)
selected_threshold = roc_df['Threshold'].iloc[2:9].values

# 각 임계값에 대해 평가 수행
for input_threshold in selected_threshold:
    y_pred = np.where(model.predict_proba(X_test)[:, 1] >= input_threshold, 1, 0)
    acc = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    print(f"정확도: {acc:.2f} | 민감도: {recall:.2f} | 정밀도: {precision:.2f} | cutoff: {input_threshold:.2f} | AUC: {roc_auc:.2f}")
    display(pd.DataFrame(confusion_matrix(y_test, y_pred)))
    print("====================================")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score
from sklearn.preprocessing import StandardScaler
from IPython.display import display

# 'mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 0과 1로 변환
df1['판정'] = df1['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

# test size를 0.2에서 0.5까지 확인
for test_size in np.arange(0.2, 0.6, 0.1):
    # 트레인 테스트 스플릿
    X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=test_size, random_state=42)

    # IQR을 사용하여 이상치 제거
    Q1 = X_train.quantile(0.25)
    Q3 = X_train.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 이상치 제거
    X_train_no_outliers = X_train[(X_train >= lower_bound) & (X_train <= upper_bound)].dropna()
    y_train_no_outliers = y_train.loc[X_train_no_outliers.index]

    # 데이터 스케일링
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_no_outliers)
    X_test_scaled = scaler.transform(X_test)

    # KNN 모델 생성 및 학습
    knn_model = KNeighborsClassifier(n_neighbors=5)  # 예시로 5개의 이웃 사용
    knn_model.fit(X_train_scaled, y_train_no_outliers)

    # 테스트 데이터에 대한 예측
    y_pred = knn_model.predict(X_test_scaled)

    # 혼동 행렬 출력
    print(f"\nTest Size: {test_size}")
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    display(pd.DataFrame(conf_matrix, columns=['Predicted 0', 'Predicted 1'], index=['Actual 0', 'Actual 1']))

    # 정확도, 민감도, 정밀도 출력
    acc = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    print(f"Accuracy: {acc:.2f} | Recall: {recall:.2f} | Precision: {precision:.2f}\n")

    # 혼동 행렬 시각화
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, roc_auc_score, auc

# 'mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 0과 1로 변환
df1['판정'] = df1['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

# 타겟 변수 설정
target = df1['판정']
test_size = 0.3
# 트레인 테스트 스플릿
X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=0.3, random_state=42)

# IQR을 사용하여 이상치 제거
Q1 = X_train.quantile(0.25)
Q3 = X_train.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 제거
X_train_no_outliers = X_train[(X_train >= lower_bound) & (X_train <= upper_bound)].dropna()
y_train_no_outliers = y_train.loc[X_train_no_outliers.index]

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_no_outliers)
X_test_scaled = scaler.transform(X_test)

# KNN 모델 생성 및 학습
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train_no_outliers)

# SVM 모델 생성 및 학습
svm_model = SVC(kernel='linear', probability=True, random_state=42)
svm_model.fit(X_train_scaled, y_train_no_outliers)

# 로지스틱 회귀 모델 생성 및 학습
logreg_model = LogisticRegression(random_state=42)
logreg_model.fit(X_train_scaled, y_train_no_outliers)

# 결정 트리 모델 생성 및 학습
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_scaled, y_train_no_outliers)

# 각 모델의 예측 확률을 얻음
y_prob_knn = knn_model.predict_proba(X_test_scaled)[:, 1]
y_prob_svm = svm_model.predict_proba(X_test_scaled)[:, 1]
y_prob_logreg = logreg_model.predict_proba(X_test_scaled)[:, 1]
y_prob_dt = dt_model.predict_proba(X_test_scaled)[:, 1]

# ROC 곡선을 그리기 위해 FPR 및 TPR 계산
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_prob_knn)
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_prob_svm)
fpr_logreg, tpr_logreg, _ = roc_curve(y_test, y_prob_logreg)
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)

# ROC 곡선의 면적(AUC) 계산
roc_auc_knn = auc(fpr_knn, tpr_knn)
roc_auc_svm = auc(fpr_svm, tpr_svm)
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)
roc_auc_dt = auc(fpr_dt, tpr_dt)

# ROC 곡선 그리기
plt.figure(figsize=(8, 6))
plt.plot(fpr_knn, tpr_knn, label=f'KNN (AUC = {roc_auc_knn:.3f})')
plt.plot(fpr_svm, tpr_svm, label=f'SVM (AUC = {roc_auc_svm:.3f})')
plt.plot(fpr_logreg, tpr_logreg, label=f'Logistic Regression (AUC = {roc_auc_logreg:.3f})')
plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC = {roc_auc_dt:.3f})')

# 랜덤 모델의 ROC 곡선 그리기
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')

plt.title('ROC Curves for Different Models with Outlier Removal')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
