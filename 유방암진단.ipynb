```python
import pandas as pd

# CSV 파일에서 데이터프레임 읽기
df1 = pd.read_csv('data.csv')

# 'func1' 함수 정의
def func1(row):
    if 'M' in row:
        return 1
    else:
        return 0

# 'diagnosis' 열에 'func1' 함수 적용하여 '판정' 열 생성
df1['판정'] = df1['diagnosis'].apply(func1)

# 결과 확인
df1.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, accuracy_score, recall_score, precision_score, confusion_matrix, roc_auc_score
from IPython.display import display

# 'mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 0과 1로 변환
df1['판정'] = df1['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

# 타겟 변수 설정
target = df1['판정']

# 트레인 테스트 스플릿
X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=0.4, random_state=42)

# 로지스틱 회귀 모델 생성 및 학습
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# ROC 곡선 생성
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
roc_df = pd.DataFrame({'FPR': fpr, 'TPR': tpr, 'Threshold': thresholds})
roc_df['Threshold'] = roc_df['Threshold'].round(2)

# AUC 계산
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

# 선택된 임계값 범위 설정 (인덱스 2~8)
selected_threshold = roc_df['Threshold'].iloc[2:9].values

# 각 임계값에 대해 평가 수행
for input_threshold in selected_threshold:
    y_pred = np.where(model.predict_proba(X_test)[:, 1] >= input_threshold, 1, 0)
    acc = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    print(f"정확도: {acc:.2f} | 민감도: {recall:.2f} | 정밀도: {precision:.2f} | cutoff: {input_threshold:.2f} | AUC: {roc_auc:.2f}")
    display(pd.DataFrame(confusion_matrix(y_test, y_pred)))
    print("====================================")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score
from sklearn.preprocessing import StandardScaler
from IPython.display import display

# 'mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 0과 1로 변환
df1['판정'] = df1['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

# 타겟 변수 설정
target = df1['판정']

# 트레인 테스트 스플릿
X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=0.4, random_state=42)

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KNN 모델 생성 및 학습
knn_model = KNeighborsClassifier(n_neighbors=5)  # 예시로 5개의 이웃 사용
knn_model.fit(X_train_scaled, y_train)

# 테스트 데이터에 대한 예측
y_pred = knn_model.predict(X_test_scaled)

# 혼동 행렬 출력
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
display(pd.DataFrame(conf_matrix, columns=['Predicted 0', 'Predicted 1'], index=['Actual 0', 'Actual 1']))

# 정확도, 민감도, 정밀도 출력
acc = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
print(f"\nAccuracy: {acc:.2f} | Recall: {recall:.2f} | Precision: {precision:.2f}\n")

# 혼동 행렬 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 'Mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 그대로 사용
target = df1['판정']

# 트레인 테스트 스플릿
X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=0.2, random_state=42)

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# PCA를 사용하여 2차원으로 축소
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# KNN 모델 생성 및 학습
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_pca, y_train)

# 결정 경계 시각화
h = .02  # mesh의 스텝 사이즈

x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = knn_model.predict(np.c_[xx.ravel(), yy.ravel()])

# 결과를 색으로 표시
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Blues, alpha=0.8)

# 훈련 데이터와 테스트 데이터를 플로팅
scatter_train = plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolor='k', marker='o')
scatter_test = plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, cmap=plt.cm.Paired, edgecolor='k', marker='x')

plt.legend([scatter_train, scatter_test], ['Train Data', 'Test Data'])
plt.title('KNN Decision Boundary Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 'Mean'이 들어간 열만 선택
mean_columns = total.filter(like='mean')

# '판정' 열을 그대로 사용
target = df1['판정']

# 트레인 테스트 스플릿
X_train, X_test, y_train, y_test = train_test_split(mean_columns, target, test_size=0.5, random_state=42)

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# PCA를 사용하여 2차원으로 축소
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# SVM 모델 생성 및 학습
svm_model = SVC(kernel='linear', C=1)  # 선형 커널 사용 (다른 커널도 사용 가능)
svm_model.fit(X_train_pca, y_train)

# 결정 경계 시각화
h = .02  # mesh의 스텝 사이즈
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])

# 결과를 색으로 표시
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Blues, alpha=0.8)

# 훈련 데이터와 테스트 데이터를 플로팅
scatter_train = plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolor='k', marker='o')
scatter_test = plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, cmap=plt.cm.Paired, edgecolor='k', marker='x')

plt.legend([scatter_train, scatter_test], ['Train Data', 'Test Data'])
plt.title('SVM Decision Boundary Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
